# Data cleaning
### removing Unnamed: 0

```python
df = df.drop(columns=["Unnamed: 0"])
```

### first we clean data brand_name how many NaN values are ther in column removing entire rows

```python
nan= df['brand_name'].isna().sum()
print("No.of NaNs in 'brand_name' column:", nan)

df_data= df[df['brand_name'].notna()].reset_index(drop=True)
print("New DataFrame shape:", df_data.shape)
```

```python
df_data
```

```python
output_file = "flipkart_EDAproject.csv"
df_data.to_csv(output_file, index=True)
```

```python
df_data.info()
```

```python
df_data.isnull().sum()
```

```python
df_data['ratings']=df_data['ratings'].fillna(value=df_data['ratings'].median())
```

### second removing " , " in price,no_of_ratings and reviews

#### price

```python
df_data["price"]=df_data["price"].str.replace(",","")
df_data["price"]=df_data["price"].astype(int)
```

#### no_of_ratings

```python
df_data["no_of_ratings"]=df_data["no_of_ratings"].str.replace(",","")
df_data["no_of_ratings"]=df_data["no_of_ratings"].astype(int)
```

#### reviews

```python
df_data["reviews"]=df_data["reviews"].str.replace(",","")
df_data["reviews"]=df_data["reviews"].astype(int)
```

```python
df_data['ratings']=df_data['ratings'].fillna(value=df_data['ratings'].median())
```

```python
output_file = "flipkart_EDAproject_CleandData.csv"
df_data.to_csv(output_file, index=True)
```

## Bivariate

#### How does the average TV price differ by brand and screen type?

```python
plots = pd.crosstab(index=df_data['brand_name'],columns=df_data['screen_type'], values=df_data['price'],aggfunc="mean")
plots.plot(kind='bar',stacked=False )
plt.title("Average Price by Brand and ScreenType")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.xticks(rotation=45) 
plt.show()
```

-the plot is visually represent to average prices by brands and screen type using scatter plot.

-highlighting how prices vary across their combinations.

#### How does TV price impact customer ratings?

#### How do TV ratings vary by screen size?

```python
size_ratings = df_data.groupby('screen_size_inch')['ratings'].mean()

plt.figure(figsize=(8, 4))
plt.plot(size_ratings.index, size_ratings.values, marker='o', linestyle='-', color='purple')

plt.xlabel("Screen Size (inches)")
plt.ylabel("Average Ratings")

plt.title("Screen Size vs. Customer Ratings")
plt.grid(True)
plt.show()
```

- The plot shows the relationship between screen size and average customer ratings for TVs.

- Customer ratings vary slightly with screen size, reaching the highest average rating at the 50-inch screen size.

- The lowest average rating appears around the 22-inch screen size, while larger screens generally maintain higher ratings.

#### Do higher-rated TVs receive more customer reviews?

```python
review_data = df_data.groupby('ratings')['no_of_ratings'].sum().reset_index()

#plt.figure(figsize=(8, 6))
plt.scatter(review_data['ratings'], review_data['no_of_ratings'], s=review_data['no_of_ratings'] /500, )
plt.xlabel("Customer Ratings")
plt.ylabel("Total Number of Reviews")
plt.title("Do Higher Ratings Lead to More Reviews?")
plt.show()
```

- The plot are exploring whether higher customer ratings result in more product reviews.

- TVs with lower ratings (around 4.5) have received the highest number of reviews.

- Higher-rated TVs (4.1–4.2) tend to have fewer total reviews, suggesting that high ratings don’t always correlate with high review counts.

## MULTIVARIATE ANALYSIS

#### How are ratings related to screen type and resolution

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_data, x='screen_type', y='ratings', hue='resolution', palette='coolwarm')

plt.xlabel("Screen Type")
plt.ylabel("Customer Ratings")
plt.title("Screen Type vs. Ratings Across Resolutions")
plt.show()
```

- The plot compares customer ratings across different screen types (LED vs QLED) and resolutions (HD Ready, Full HD, Ultra HD 4K).

#### How do screen size, price, and ratings interact

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_data['screen_size_inch'], df_data['ratings'], df_data['price'], c=df_data['launch_year'], cmap='viridis')
ax.set_xlabel('Screen Size (cm)')
ax.set_ylabel('Ratings')
ax.set_zlabel('Price')
ax.set_title('Screen Size, Ratings, and Launch Year vs Price')
plt.show()
```

TVs with bigger screens and better ratings usually price more.
Newer TV models, launched in recent years, are also priced much higher than older ones.

#### How do screen size, price, and ratings interact?

```python
sns.scatterplot(x='screen_size_cm', y='price', hue='ratings',data=df_data, palette='viridis', sizes=(20, 200))
plt.title('Screen Size vs Price (Hue: Ratings, Size: No. of Ratings)')
plt.show()
```

TVs with larger screen sizes tend to have higher prices,but the rattings is verry less compare to less prices and less screen_sizes ,less expensive TVs

#### What are the correlations among numerical variables?

```python
corr = df_data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".1f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

- Price and screen size show a positive correlation, confirming that larger TVs are generally more expensive

- No. of ratings and reviews are highly correlated, suggesting that popular products attract more reviews.
